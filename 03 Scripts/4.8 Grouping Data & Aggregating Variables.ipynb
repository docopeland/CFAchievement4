```python
#importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#setting up data frames
path = r'/Users/docopeland/04 Instacart Basket Analysis'
ords_prods_merge = pd.read_pickle(os.path.join(path,'02 Data','prepared data','ords_prods_freq.pkl'))
#subset of ords_prods_merge
df = ords_prods_merge[:1000000]
```

```python
df.head(10)
```

```python
#grouping by product_name column
df.groupby('product_name')
```

```python
#group the data by department_id and return mean
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#group the data by department_id and return mean, without .agg function
df.groupby('department_id')['order_number'].mean()
```

```python
#group the data by department_id and return mean, min, max
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#create a new column that provides the max number of orders per user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(20)
```

```python
#setting up my loyalty flags
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking out the values that were created
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
