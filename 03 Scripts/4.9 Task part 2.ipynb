```python
#import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot
import seaborn as sns
import scipy
```

```python
#import files
path = r'/Users/docopeland/04 Instacart Basket Analysis'
customers = pd.read_pickle(os.path.join(path,'02 Data','prepared data','cust_ord_prods.pkl'))
```

```python
#bar chart
bar1 = customers['order_dow'].value_counts().sort_index().plot.bar()
```

```python
#histogram
hist1 = customers['prices'].plot.hist(bins = 25)
```

```python
#scatterplot
scatt1= sns.scatterplot(x = 'prices', y = 'prices',data = customers)
```

```python
#creating a sample with a seed of 4 with 70/30
np.random.seed(4)
dev = np.random.rand(len(customers)) <= 0.7
small = customers[~dev]
```

```python
#line plot
line1 = sns.lineplot(data = small[['order_dow','prices']], x = 'order_dow',y = 'prices')
```

```python
#order_hour_of_day histogram
#this histogram shows how many orders were done per hour, the most popular hours are 10am, 11am, 2pm, and 3pm.
histHOD = customers['order_hour_of_day'].plot.hist(bins = 25)
```

```python
#the above histogram shows how many orders were done per hour, the most popular hours are 10am, 11am, 2pm, and 3pm.
```

```python
#loyalty flag bar
barLoyalty = customers['loyalty_flag'].value_counts().plot.bar()
```

```python
#regular customers have more orders than loyal customers
```

```python
#difference in expenditure by hour of day
lineExpHOD = sns.lineplot(data = small[['order_hour_of_day','prices']], x = 'order_hour_of_day', y = 'prices')
```

```python
#connection between age and number of dependents
lineAgeDep = sns.lineplot(data = small[['Age', 'dependenants']], x = 'Age', y = 'dependenants')
```

```python
#there doesn't seem to be any sort of connection
```

```python
#connection between age and spending power (income)
scattAgeSpending = sns.scatterplot(x = 'Age', y = 'income', data = small)
```

```python
#there seems to be no connection between age and spending power (income), people at all ages have all different incomes
```
