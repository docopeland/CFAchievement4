```python
#question 6, get more info about breakfast 
#used a for loop to find out what department breakfast was
val = 0
for i in list(data_dict):
    if(data_dict.get(i) == {'department': 'breakfast'}):
        val = int(i)
# find all the info for depatment 
df_breakfast = df_products.loc[df_products['department_id'] == val]
df_breakfast
```

```python
#question 7, dinner parties so departments: alcohol, deli, beverages, and meat/seafood
#find out which departments these are in
dinnerparty = ['alcohol','deli','beverages','meat seafood']
deptno = []
for i in list(data_dict):
    for j in range(0,len(dinnerparty)):
        if data_dict.get(i) == {'department' : dinnerparty[j]}:
            deptno.append(int(i))
#get all the info from those departments
df_dinnerparty = df_products.loc[df_products['department_id'].isin(deptno)]
df_dinnerparty
```
